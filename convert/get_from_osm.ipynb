```python
import overpy

api = overpy.Overpass()
# Define the Overpass API query
overpass_query = """
[out:json];
(
    node["tourism"~"^({})$"]({bbox});
    node["leisure"~"^({})$"]({bbox});
    node["amenity"~"^({})$"]({bbox});
    node["historic"~"^({})$"]({bbox});
);
out center;
""".format("|".join(tourism_tags), "|".join(leisure_tags), "|".join(amenity_tags),"|".join(historic_tags), bbox=",".join(str(x) for x in bbox))

print(overpass_query)
# Send the query to the Overpass API and get the results
response = api.query(overpass_query)
```

```python
print(len(response.nodes))
print(response.nodes[0])

print(overpass_query)
```

```python
import json

with open("../data/raw/osm.json") as f:
    data = json.load(f)

elements = data["elements"]

new_elements = []
for e in elements:
    tag = e['tags']
    el = {k: v for k, v in e.items() if k != 'tags'}
    new_elements.append({**el, **tag})

new_elements
```

```python
with open("new_elements.json", "w") as f:
    json.dump(new_elements, f)
```

```python
import polars as pl

# Convert the JSON data to Arrow table
df = pl.read_json("./new_elements.json")

df
```
